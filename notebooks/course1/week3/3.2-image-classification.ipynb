```python
from transformers import pipeline
from PIL import Image
import requests
from io import BytesIO
```

## 1. Create Image Classification Pipeline

```python
classifier = pipeline("image-classification", model="google/vit-base-patch16-224", device="cpu")
print(f"Model loaded: {classifier.model.config.model_type}")
```

## 2. Classify an Image

```python
# Load image from URL (using HuggingFace docs which doesn't block requests)
url = "https://huggingface.co/datasets/huggingface/documentation-images/resolve/main/pipeline-cat-chonk.jpeg"
response = requests.get(url)
image = Image.open(BytesIO(response.content))

# Classify
results = classifier(image)
for r in results[:3]:
    print(f"{r['label']}: {r['score']:.3f}")
```
