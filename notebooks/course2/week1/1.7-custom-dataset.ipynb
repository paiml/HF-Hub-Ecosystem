## 1. Generator-Based Dataset

```python
def data_generator():
    """Generate sample data."""
    samples = [
        {"text": "This is great!", "label": 1},
        {"text": "This is bad.", "label": 0},
        {"text": "Amazing product!", "label": 1},
    ]
    for sample in samples:
        yield sample

ds = Dataset.from_generator(data_generator)
print(f"Dataset: {ds}")
```

## 2. Adding Metadata

```python
# Add text length column
ds = ds.map(lambda x: {"length": len(x["text"])})
print(ds[0])
```
