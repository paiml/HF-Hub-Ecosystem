````python
from transformers import TrainingArguments
````

## 1. Configure Logging

````python
training_args = TrainingArguments(
    output_dir="./results",
    num_train_epochs=3,
    per_device_train_batch_size=8,
    logging_dir="./logs",
    logging_steps=10,
    logging_strategy="steps",
    report_to=["tensorboard"],
)

print(f"Logging to: {training_args.logging_dir}")
print(f"Report to: {training_args.report_to}")
````

## 2. TensorBoard Integration

To view logs, run:
```bash
tensorboard --logdir ./logs
```

````python
# Note: In a real training loop, metrics are automatically logged
# The Trainer handles this when report_to=["tensorboard"] is set
print("TensorBoard logging configured")
````
